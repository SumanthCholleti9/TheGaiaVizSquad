```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
import warnings
warnings.filterwarnings("ignore")
```

```python
countriesWithLatAndLongDF = pd.read_csv("country-coord.csv")
countriesWithLatAndLongDFReduced = countriesWithLatAndLongDF[["Country", "Latitude (average)", "Longitude (average)"]]
countriesWithLatAndLongDFReduced = countriesWithLatAndLongDFReduced.reset_index()
countriesWithLatAndLongDFReduced['index'] += 1
countriesWithLatAndLongDFReduced = countriesWithLatAndLongDFReduced.rename(columns={'Country': 'GeoAreaName', "Latitude (average)":'translate_y',  "Longitude (average)":'translate_x', 'index' : 'np_node_id'})
numberofCountryPins = countriesWithLatAndLongDFReduced.shape[0]
countriesWithLatAndLongDFReduced
```

```python
nodeDFforGaiaVizBase = pd.read_csv("node.csv")
nodeDFforGaiaViz = nodeDFforGaiaVizBase.drop(columns = ["translate_x","translate_y"])

defaultValuesForNodeDF = nodeDFforGaiaViz.mode().dropna()
newNodeDF = defaultValuesForNodeDF._append([defaultValuesForNodeDF.loc[0].copy()] * (numberofCountryPins - 1), ignore_index=True)
newNodeDF = newNodeDF.astype(nodeDFforGaiaViz.dtypes.to_dict())

newNodeDF["np_node_id"] = pd.Series(range(1, numberofCountryPins+1))
newNodeDF["np_data_id"] = newNodeDF["np_node_id"]
newNodeDF["np_tag_id"] = newNodeDF["np_node_id"]
newNodeDF["record_id"] = newNodeDF["np_node_id"]
newNodeDF["scale_x"] = 1.474016
newNodeDF["scale_y"] = 1.477102
newNodeDF["scale_z"] = 1.474524

newNodeDF['tag_offset_y'] = 0
newNodeDF['tag_offset_z'] = 10

newNodeDF['segments_x'] = 1
newNodeDF['segments_x'] = 1
newNodeDF['segments_x'] = 1

newNodeDF['ratio'] = 0.03
newNodeDF['np_geometry_id'] = 19
newNodeDF['np_topo_id'] = 6
newNodeDF['color_r'] = 194
newNodeDF['color_g'] = 30
newNodeDF['color_b'] = 121

newNodeDF
```

```python
baseNodeDFWithCountryPins = pd.merge(countriesWithLatAndLongDFReduced, newNodeDF, on = "np_node_id")

baseNodeDFWithCountryPins = baseNodeDFWithCountryPins.drop(columns = "GeoAreaName")
baseNodeDFWithCountryPins = baseNodeDFWithCountryPins.reindex(columns=nodeDFforGaiaVizBase.columns)
```

```python
#baseNodeDFWithCountryPins.to_csv("basePinsWithCountryNames.csv", index = False)
```

```python
baseNodeDFWithCountryPins
```

```python
tagDFforGaiaVizBase = pd.read_csv("tag.csv") #ForColumnNames
tagDF = countriesWithLatAndLongDFReduced.rename(columns= {"GeoAreaName" : "title", "np_node_id" : "np_tag_id" }).drop(columns=["translate_y", "translate_x"])
tagDF["record_id"] = tagDF["np_tag_id"]
tagDF["table_id"] = 0
tagDF["description"] = ""
tagDF = tagDF.reindex(columns=tagDFforGaiaVizBase.columns)
```
