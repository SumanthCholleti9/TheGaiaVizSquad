## Data Cleaning

```python
import pandas as pd
```

```python
# Hardcoding some columns to be removed 
cols_to_drop = ['Unnamed: 0', 'ReleaseStatus', 'ReleaseName', 'isDSDSeries', 'ValueType', 'FootNote', 'Source', 'Reporting Type', 'Observation Status', 'BasePeriod']

for i in cols_to_drop:
    dfg1.drop(i, inplace=True, axis=1)
```

```python
dfg1.shape
```

```python
# Counting the number of null values in each column

cols_to_drop = []

# Columns that have % values as NULL 
null_counts = dfg1.isnull().sum()
for each_col, each_col_null_count in null_counts.items():
    if ((each_col_null_count/len(dfg1))*100) == 100:
        print(each_col, each_col_null_count)
        cols_to_drop.append(each_col)
```

```python
len(cols_to_drop)
```

```python
for i in cols_to_drop:
    dfg1.drop(i, inplace=True, axis=1)
```

```python
dfg1.shape
```

```python
null_counts = dfg1.isnull().sum()

for each_col, each_col_null_count in null_counts.items():
    print(each_col, "\t", round((each_col_null_count/len(dfg1))*100))
```
